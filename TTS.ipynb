```python
import fitz  # PyMuPDF

# Load PDF
doc = fitz.open("my_file.pdf")

# Extract all text
full_text = ""
for page in doc:
    full_text += page.get_text()

# Save to text file
with open("my_file_qna.txt", "w", encoding="utf-8") as f:
    f.write(full_text)
```

```python
with open("my_file_qna.txt", "r", encoding="utf-8") as file:
    text = file.read()

# Use gTTS to convert to audio
tts = gTTS(text=text, lang='en', slow=True)
tts.save("my_audio.mp3")

# Play audio in Colab
Audio("my_audio.mp3")

```
